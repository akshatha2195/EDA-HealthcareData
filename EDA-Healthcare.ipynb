**Missing Data - Initial Intuition**


```python
# Creating a DataFrame with missing values percentage
missing = pd.DataFrame({'Column': df.columns, 'MissingPercent': (df.isnull().sum() * 100 / df.shape[0])})

# Plotting
plt.figure(figsize=(16,5))
ax = sns.pointplot(x='Column', y='MissingPercent', data=missing)

plt.xticks(rotation=90, fontsize=10)
plt.title("Percentage of Missing Values", fontsize=40)
plt.ylabel("PERCENTAGE", fontsize=20)
plt.grid()
plt.show()
```

Here, we don't have any missing data.

**General Thumb Rules:**

For features with less missing values- can use regression to predict the missing values or fill with the mean of the values present, depending on the feature.
For features with very high number of missing values- it is better to drop those columns as they give very less insight on analysis.
As there's no thumb rule on what criteria do we delete the columns with high number of missing values, but generally you can delete the columns, if you have more than 30-40% of missing values.

BiVariate Anlaysis

```python
data_target0=df.loc[df["Noshow"]==0]
data_target1=df.loc[df["Noshow"]==1]
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30

    # Check if hue is provided, else use an empty Series
    temp = pd.Series(df[hue]) if hue else pd.Series([])

    # Dynamic figure width
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig, ax = plt.subplots(figsize=(max(width, 10), 8))  # Ensure minimum width

    # Apply log scale only if no zero values exist
    if (df[col].value_counts() > 0).all():
        plt.yscale('log')

    plt.title(title)
    ax = sns.countplot(data=df, x=col, order=df[col].value_counts().index, hue=hue, palette='bright')

    plt.show()
```

```python
uniplot(df, col='Noshow', title='Distribution of NoShow', hue='Gender')
```

```python
uniplot(data_target0,col='Hypertension',title='Distribution of Gender for NoShow Customers',hue='Gender')
```

```python
uniplot(data_target1,col='Age_group',title='Distribution of Age for NoShow Customers',hue='Gender')
```
